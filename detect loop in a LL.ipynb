```python
class Node:

	def __init__(self):
		self.data = 0
		self.next = None
		self.flag = 0


def push(head_ref, new_data):
	''' allocate node '''
	new_node = Node()

	''' put in the data '''
	new_node.data = new_data

	new_node.flag = 0

	''' link the old list of the new node '''
	new_node.next = (head_ref)

	''' move the head to point to the new node '''
	(head_ref) = new_node
	return head_ref



def detectLoop(h):

	while (h != None):
		if (h.flag == 1):
			return True

		h.flag = 1
		h = h.next
	return False


''' Driver program to test above function'''
if __name__ == '__main__':

	''' Start with the empty list '''
	head = None

	head = push(head, 20)
	head = push(head, 4)
	head = push(head, 15)
	head = push(head, 10)

	''' Create a loop for testing '''
	head.next.next.next.next = head

	if (detectLoop(head)):
		print("Loop Found")
	else:
		print("No Loop")
#Time complexity=O(n)
#Space complexity=O(1)


```
