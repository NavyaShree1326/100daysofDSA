```python
class linkedList: 
	def __init__(self): 
		self.head = None
		self.tail = None

	def insert(self, data): 
		if self.head is None: 
			self.head = Node(data) 
			self.tail = self.head 
		else: 
			self.tail.next = Node(data) 
			self.tail = self.tail.next


class Node: 
	def __init__(self, data): 
		self.data = data 
		self.next = None


def findIntersection(head1, head2): 
	hashmap = {} 

	while(head1 != None): 
		data = head1.data 
		if(data not in hashmap.keys()): 
			hashmap[data] = 1
		head1 = head1.next
	ans = linkedList() 
	while(head2 != None): 
		data = head2.data 
		if(data in hashmap.keys()): 
			ans.insert(data) 
		head2 = head2.next
	return ans.head 


def union(head1, head2): 
	hashmap = {} 
	while(head1 != None): 
		data = head1.data 
		if(data not in hashmap.keys()): 
			hashmap[data] = 1
		head1 = head1.next

	while(head2 != None): 
		data = head2.data 
		if(data not in hashmap.keys()): 
			hashmap[data] = 1
		head2 = head2.next
	ans = linkedList() 
	for key, value in hashmap.items(): 
		ans.insert(key) 

	return ans.head 

def printList(head): 
	while head: 
		print(head.data, end=' ') 
		head = head.next
	print() 


if __name__ == '__main__': 

	ll1 = linkedList() 
	ll1.insert(1) 
	ll1.insert(2) 
	ll1.insert(3) 
	ll1.insert(4) 
	ll1.insert(5) 

	ll2 = linkedList() 
	ll2.insert(1) 
	ll2.insert(3) 
	ll2.insert(5) 
	ll2.insert(6) 

	print("First list is ") 
	printList(ll1.head) 

	print("Second list is ") 
	printList(ll2.head) 

	print("Intersection list is") 
	printList(findIntersection(ll1.head, ll2.head)) 

	print("Union list is ") 
	printList(union(ll1.head, ll2.head)) 

#Time Complexity=O(m+n)
#Space Complexity=O(m+n)

```
